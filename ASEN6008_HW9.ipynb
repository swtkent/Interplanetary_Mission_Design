# 3

```python
Mr = 1.524*AU_KM
Pr = 39.482*AU_KM

Mv = np.sqrt(MU_SUN_KM/(Mr))  # Velocity of Earth

atr = (Mr+Pr)/2    # Transfer semi-major axis

vtrp = np.sqrt(2*MU_SUN_KM/Mr - MU_SUN_KM/atr)    # Velocity at periapsis of the transfer

vInf = vtrp - Mv  # V infinity at Earth

print(f'Vinf at Mars: {vInf:.3f} km/s')

## Venus to Neptune
Vr = 0.723*AU_KM
Nr = 30.069*AU_KM

Vv = np.sqrt(MU_SUN_KM/(Vr))  # Velocity of Earth


atr = (Vr+Nr)/2    # Transfer semi-major axis

vtrp = np.sqrt(2*MU_SUN_KM/Vr - MU_SUN_KM/atr)    # Velocity at periapsis of the transfer

vInf = vtrp - Vv  # V infinity at Earth

print(f'Vinf at Venus: {vInf:.3f} km/s')
```
